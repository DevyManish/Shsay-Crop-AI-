```python
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df

num_cols = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
df_clean = remove_outliers(df, num_cols)
df_clean.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_clean.drop('label', axis=1)
y = df_clean['label']
```

```python
X.head(5)
```

```python
y.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    'DecisionTreeClassifier': DecisionTreeClassifier(criterion="entropy",random_state=2,max_depth=5),
    'GaussianNB': GaussianNB(),
    'SVC': SVC(gamma='auto'),
    'LogisticRegression': LogisticRegression(random_state=42),
    'RandomForestClassifier': RandomForestClassifier(n_estimators=20, random_state=0)
}
```

```python
from sklearn.metrics import accuracy_score

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = {
        'accuracy': acc,
    }
    print(f"{name} Accuracy: {acc:.4f}")
    print("-" * 50)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

rf_grid = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=5, scoring='accuracy', n_jobs=-1)
rf_grid.fit(X_train, y_train)
print("Best RandomForestClassifier Params:", rf_grid.best_params_)
print("Best RandomForestClassifier Score:", rf_grid.best_score_)
```

```python
gnb_params = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6]
}

gnb_grid = GridSearchCV(GaussianNB(), gnb_params, cv=5, scoring='accuracy', n_jobs=-1)
gnb_grid.fit(X_train, y_train)
print("Best GaussianNB Params:", gnb_grid.best_params_)
print("Best GaussianNB Score:", gnb_grid.best_score_)
```

```python
import pickle as pkl

with open('best_randomforest_model.pkl', 'wb') as rf_file:
    pkl.dump(rf_grid.best_estimator_, rf_file)

with open('best_gaussiannb_model.pkl', 'wb') as gnb_file:
    pkl.dump(gnb_grid.best_estimator_, gnb_file)
```

```python
from sklearn.metrics import accuracy_score, classification_report


with open('best_randomforest_model.pkl', 'rb') as rf_file:
    rf_model = pkl.load(rf_file)
with open('best_gaussiannb_model.pkl', 'rb') as gnb_file:
    gnb_model = pkl.load(gnb_file)
```

```python
rf_pred = rf_model.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
print('RandomForestClassifier Test Accuracy:', rf_acc)
print('RandomForestClassifier Classification Report:')
print(classification_report(y_test, rf_pred))
```

```python
gnb_pred = gnb_model.predict(X_test)
gnb_acc = accuracy_score(y_test, gnb_pred)
print('GaussianNB Test Accuracy:', gnb_acc)
print('GaussianNB Classification Report:')
print(classification_report(y_test, gnb_pred))
```
